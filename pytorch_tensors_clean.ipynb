```python
w = np.array([0.5])
b = np.array([2])
xx = np.array([0.5])#np.arange(0,1.5,.5)
```

```python
xx_t = torch.from_numpy(xx)
w_t = torch.from_numpy(w)
b_t = torch.from_numpy(b)
```

```python
w_t.requires_grad
```

```python
w_t.requires_grad_(True)
```

```python
b_t.requires_grad_(True)
```

```python
dtype = torch.float64
```

```python
b_t = b_t.type(dtype)
```

```python
b_t.requires_grad_(True)
```

```python
def fun(x,ystar):
    y = torch.exp(w_t*x+b_t)
    print(y)
    return torch.sum((y-ystar)**2)

ystar_t = torch.randn_like(xx_t)
l_t = fun(xx_t,ystar_t)
```

```python
l_t
```

```python
l_t.requires_grad
```

```python
print(w_t.grad)
```

```python
l_t.backward()
```

```python
print(w_t.grad)
print(b_t.grad)
```

```python
yy_t = torch.exp(w_t*xx_t+b_t)
print(torch.sum(2*(yy_t-ystar_t)*yy_t*xx_t))
print(torch.sum(2*(yy_t-ystar_t)*yy_t))
```

```python
l_t = fun(xx_t,ystar_t)
l_t.backward()
```

```python
print(w_t.grad)
print(b_t.grad)
```

```python
l_t.backward()
```

```python
w_t.grad.data.zero_()
b_t.grad.data.zero_()
l_t = fun(xx_t,ystar_t)
l_t.backward(retain_graph=True)
l_t.backward()
print(w_t.grad)
print(b_t.grad)
```
